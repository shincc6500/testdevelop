```python
from bs4 import BeautifulSoup
import xmltodict
import json

import requests
import pandas as pd
import numpy as np


#### load
html = requests.get('https://soil.rda.go.kr/nutri/introMineral.do').text
html_soup = BeautifulSoup(html, 'html.parser')

#### contents
contents = html_soup.select("dl.article_c") # 작물 단위 테이블


#### 컬럼 모음
df_col_list = ["작물", "재배법", "생육시기", "부위",
            "질소", "인산", "칼륨", "칼슘", "마그네슘",
            "철", "망간", "아연", "붕소", "규소",]

df = pd.DataFrame(columns=df_col_list)
print(df_col_list)

#### 값 추출
for content in contents:
    print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
    ## 컬럼
    thead = content.select("thead")[0] 
    col_list = thead.text.split()
    n_idx = col_list.index("N")
    n_kr_idx = col_list.index("질소")
    
    crop = content.select("dt")[0]
    crop = crop.text.split() # 작물 이름
    
    col_list = ["작물"] + col_list[:n_idx] + col_list[n_kr_idx:n_kr_idx+(n_kr_idx-n_idx)]
    
    ## 영양 값
    val_list = content.select("tbody")[0].text.split() # 필요 영양 값
    
    len_values = len(val_list)
    len_02 = int(len_values/2)
    len_03 = int(len_values/3)
    len_23 = int(len_values*2/3)
    if len_values <= 13: val_list = [crop + val_list]
    elif len_values <= 26: val_list = [crop + val_list[:len_02], crop + val_list[len_02:]]
    elif len_values <= 39: val_list = [crop + val_list[:len_03], crop + val_list[len_03:len_23], crop + val_list[len_23:]]
    else: print("ㅡㅡ len error ㅡㅡ")
    
    try:
        df = pd.concat([df, pd.DataFrame(data = val_list, columns=col_list)], ignore_index=True)
    except:
        print("len_values", len_values)
        print(len(val_list), val_list)
        print("2:", len_02, "3:", len_03)
        
        # break
    
    # print(df)
    # break
    
    
# print(df)
    
#### save
# wf = open('./nutri_03.txt', 'w')
# print(content_w, file = wf)
# count_01 += 1

# if count_01 == 3:
#     wf.close()
#     break


# print(type())
```
